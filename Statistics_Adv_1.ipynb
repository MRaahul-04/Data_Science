#Statistics Part 1

Here are simple, easy-to-understand answers for each of your 23 statistics questions:

---

### 1. **What is a random variable in probability theory?**
- A **random variable** is a variable that takes different values based on the outcome of a random experiment (like tossing a coin or rolling a die).

---

### 2. **What are the types of random variables?**
- **Discrete random variable**: Takes specific, countable values (e.g., 1, 2, 3).
- **Continuous random variable**: Takes any value within a range (e.g., height, temperature).

---

### 3. **What is the difference between discrete and continuous distributions?**
- **Discrete distribution** deals with countable outcomes.
- **Continuous distribution** deals with measurable outcomes over an interval.

---

### 4. **What are probability distribution functions (PDF)?**
- A **PDF** shows the likelihood (probability) of each possible outcome for a random variable.

---

### 5. **How do cumulative distribution functions (CDF) differ from PDFs?**
- **PDF** shows the probability of a specific value.
- **CDF** shows the probability that the value is **less than or equal to** a certain value.

---

### 6. **What is a discrete uniform distribution?**
- Every outcome has **equal probability**. Example: rolling a fair die (1 to 6).

---

### 7. **What are the key properties of a Bernoulli distribution?**
- Only **two outcomes**: success (1) or failure (0).
- Has **one trial**, with probability `p` of success.

---

### 8. **What is the binomial distribution, and how is it used in probability?**
- Describes the number of **successes in `n` Bernoulli trials**.
- Example: Flipping a coin 10 times, counting heads.

---

### 9. **What is the Poisson distribution and where is it applied?**
- Used for **counting events over time or space**.
- Example: Number of emails per hour.

---

### 10. **What is a continuous uniform distribution?**
- All values in a range are **equally likely**.
- Example: Random time between 1 to 5 minutes.

---

### 11. **What are the characteristics of a normal distribution?**
- Bell-shaped curve.
- Symmetrical around the **mean**.
- Mean = Median = Mode.

---

### 12. **What is the standard normal distribution, and why is it important?**
- Normal distribution with **mean = 0** and **std deviation = 1**.
- Used to calculate **Z-scores** and probabilities.

---

### 13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
- It says that the **sampling distribution** of the mean becomes **normal**, even if the original data is not, when sample size is large enough.
- Very important for **confidence intervals and hypothesis testing**.

---

### 14. **How does the Central Limit Theorem relate to the normal distribution?**
- Thanks to CLT, we can use the **normal distribution to analyze sample means**, even if the population is not normal.

---

### 15. **What is the application of Z statistics in hypothesis testing?**
Z-statistics are used to:
- **Test sample means** against population means.
- **Compare proportions** between groups.
- Used when population standard deviation is known.

---

### 16. **How do you calculate a Z-score, and what does it represent?**
- X = value, μ = mean, σ = std deviation.
- Tells how **far a value is from the mean**, in standard deviations.

---

### 17. **What are point estimates and interval estimates in statistics?**
- **Point estimate**: A single value estimate (e.g., sample mean).
- **Interval estimate**: A range (like a confidence interval) where the true value is likely to fall.

---

### 18. **What is the significance of confidence intervals in statistical analysis?**
- Confidence intervals show a **range of values** where we expect the population parameter to be, with a certain **level of confidence** (e.g., 95%).

---

### 19. **What is the relationship between a Z-score and a confidence interval?**
Z-scores define the **range** of the confidence interval. For example:
- 95% CI → Z = ±1.96
- 99% CI → Z = ±2.58

---

### 20. **How are Z-scores used to compare different distributions?**
- Z-scores **standardize** data, allowing comparisons between different distributions by showing how far values are from their respective means.

---

### 21. **What are the assumptions for applying the Central Limit Theorem?**
- Random sampling
- Independent samples
- Sample size is **large enough** (usually n ≥ 30)

---

### 22. **What is the concept of expected value in a probability distribution?**
- Expected value is the **average outcome** if the experiment is repeated many times.

---

### 23. **How does a probability distribution relate to the expected outcome of a random variable?**
- The distribution **describes how likely each value is**, and the expected outcome is the **weighted average** of these values.

---

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    x = [0, 1]
    prob = [1-p, p]
    return prob, x


x, prob = bernoulli_pdf(0.7)

plt.bar(x, prob, tick_label=["Failure (0)", "Success (1)"])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Distribution Function (PDF)')
plt.show()
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(1, 7, 1000)

counts, bin_edges = np.histogram(data, bins=6, density=True)
cdf = np.cumsum(counts) * np.diff(bin_edges)[0]  # Compute the CDF

plt.step(bin_edges[1:], cdf, where='mid', label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution')
plt.show()
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=1, size=1000)

sample_means = []
for _ in range(1000):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('CLT: Distribution of Sample Means (Non-Normal Data)')
plt.show()
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

data = [75, 88, 92, 85, 78, 90, 89, 95, 86, 83]

mean = np.mean(data)
std_err = stats.sem(data)

confidence_level = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_critical * std_err

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("Mean = ", mean)
print("SE = ", std_err)
print("Z = ", z_critical)
print("ME = ", margin_of_error)
print(f"Confidence Interval for Mean: ({lower_bound: .2f}, {upper_bound: .2f})")
```

23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 10, 1000)

mean = np.mean(data)
std = np.std(data)

normalized_data = (data - mean) / std

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, label='Original Data')
plt.title('Original Data')

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, label='Normalized Data', color='orange')
plt.title('Z-Score Normalized Data')

plt.show()
```
