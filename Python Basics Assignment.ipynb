###6. Write a function to demonstrate the behavior of mutable and immutable arguments.

````python
def new_list(list):
  list.append(4)

def new_str(s):
  s = s + "Omega"

list = [1, 2, 3]
s = "Alpha"

new_list(list)
print("New list: ", list) #Mutable

new_str(s)
print("New string: ", s)  #Immutable
````

###10. Write code to demonstrate type casting with list elements.

````python
list = [1, 2, 34.5, 45, "56", 2.0]

print(f"This is the list before type casting: - {list}\n")

for i in range(len(list)):
  list[i] = str(list[i])
print(f"Type casting list to string: {list}")

for i in range(len(list)):
  list[i] = int(float(list[i]))  #[To avoid error while type casting directly from str to int, need to convert them into float first. ]
print(f"Type casting list to integer: {list}")

for i in range(len(list)):
  list[i] = float(list[i])
print(f"Type casting list to float: {list}")
````

###13. Write a Python program to find the sum of all even numbers between 1 and 50.

````python
sum = 0

for i in range(1, 51):
  if i % 2 == 0:
    sum += i
print("Sum of all even b/w 1 and 50:- ", sum)
````

###14. Write a program to reverse a string using a while loop.

````python
a = str(input("Enter your Word: "))
i = len(a) - 1

while i >= 0:
  print(a[i], end='')
  i = i - 1
````

###15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

````python
a = int(input("Enter a nummber: "))
f = 1

for i in range(1, a+1):
  f = f * i
  i += 1

print(f"factorial of {a} is = ", f)
````
